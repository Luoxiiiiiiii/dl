```python
#另一个较好的方案

interval = [0,5,10,15,20,25,30,35,40,45,60,90]#区间
width = [5,5,5,5,5,5,5,5,5,15,30,60]#组距
quantity = [836,2737,3723,3926,3596,1438,3273,642,824,613,215,47]#频数

plt.figure(figsize = (20, 8), dpi = 80)

colors = [
    'red', 'orange', 'yellow', 'green', 'blue', "black", 'purple',
    'pink', 'cyan', 'magenta', 'lime', 'teal', 'maroon'
]

plt.bar(
    x = interval, 
    height = quantity, 
    align='edge', #条形边缘与x位置对齐（左边缘对齐起始值）
    width = width, 
    color = colors #可以设置每一条的颜色不同
)

plt.xticks(interval + [150], fontsize = 15)
plt.yticks(fontsize = 15)
plt.ylabel("频数", fontsize = 20)
plt.grid(alpha = 0.5)
plt.show()
```
