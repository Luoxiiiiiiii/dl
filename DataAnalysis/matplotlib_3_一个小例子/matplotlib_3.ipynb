```python
'''
列表a表示10点到12点的每一分钟的气温,如何绘制折线图观察每分钟气温的变化情况?
a= [random.randint(20,35) for i in range(120)]
'''
#random.randint(a, b)会生成一个随机整数N，满足a ≤ N ≤ b

#不考虑时间用10：10这样的格式表示了，就是用1-120来表示吧
```

```python
import random
from matplotlib import pyplot as plt
a=[random.randint(20,35) for i in range(120)]
b=range(1,121)
fig = plt.figure(figsize=(25,8),dpi=80)
#fig= plt.figure(figsize(20,8))是错误的
#xticks_labels = range(1,121,5)
#plt.xticks(xticks_labels)
plt.plot(b,a)
plt.show()
plt.savefig("./气温变化图.png")
#plt.figsave('./气温变化图.png')是错误的
```
