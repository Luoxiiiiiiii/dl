```python
'''
列表a表示10点到12点的每一分钟的气温,如何绘制折线图观察每分钟气温的变化情况?
a= [random.randint(20,35) for i in range(120)]
'''
#random.randint(a, b)会生成一个随机整数N，满足a ≤ N ≤ b

#不考虑时间用10：10这样的格式表示了，就是用1-120来表示吧
```

```python
import random
from matplotlib import pyplot as plt
a=[random.randint(20,35) for i in range(120)]
b=range(1,121)
fig = plt.figure(figsize=(25,8),dpi=80)
#fig= plt.figure(figsize(20,8))是错误的
#xticks_labels = range(1,121,5)
#plt.xticks(xticks_labels)
plt.plot(b,a)
plt.show()
plt.savefig("./气温变化图.png")
#plt.figsave('./气温变化图.png')是错误的
```

```python
#希望将字符串传入xticks
a = [random.randint(20,35) for i in range(120)]
x = range(120)
fig = plt.figure(figsize=(23,8),dpi = 80)
_x = list(x)[::10]
#这个代码的目的是让标签间隔显示
print(type(x))
#注意，range()的类型不是列表，而是特殊的range类型，用list强制列表转换成列表用来取间隔
#注意，range类型也是可迭代对象，可以用在for循环中
_xticks_labels = ["hello,{}".format(i) for i in _x]
#format()是字符串的内置方法，将需要填入的内容用{}占位
plt.xticks(_x,_xticks_labels)
#xticks的基本语法：plt.xticks(ticks=None,labels=None)
#None的意思是可选，可以没有，且ticks和labels参数必须一一对应，双射
#ticks和labels参数可以是列表或者元组或者range，ticks指定刻度的位置，labels指定刻度的标签
plt.plot(x,a)
plt.show()
```

```python
a = [random.randint(20,35) for i in range(120)]
min = 60
n = 2
x = range(min*n)
fig = plt.figure(figsize=(23,8),dpi = 80)

_xticks_labels = ["10点{}分".format(i) for i in range(1,min)]
#_xticks_labels_11 = ["11点{}分".format(i) for i in range(0,min)]
#_xticks_lables = _xticks_labels_10 + _xticks_labels_11 + ["12点00分"]
_xticks_labels += ["11点{}分".format(i) for i in range(0,min)] + ["12点00分"]
#可以直接用 +=
plt.xticks(ticks = list(x)[::3]+[120], labels = _xticks_labels[::3]+["12点00分"], rotation = 270, fontname = 'SimSun')
#rotation是旋转的度数
plt.plot(x,a)

#显示x轴和y轴的标签以及显示标题
plt.xlabel("时间", fontname = 'SimSun')
plt.ylabel("温度", fontname = 'SimSun')
plt.title("10点到12点每分钟的气温变化情况", fontname = 'SimSun')

plt.show()
```
