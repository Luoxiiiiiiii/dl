```python
from matplotlib import pyplot as plt
import random

#利用config设置全局属性
#config这个单词可以修改的
#Microsoft Yahei微软雅黑既有英文也有中文
#这样就实现了中文显示哈哈哈

#config = {
#    "font.family": 'Microsoft Yahei' ,
#}

"""
font.family支持字体回退机制，支持输入一个列表，绘制文本时,
会从列表中依次尝试这些字体，比如可以先设置新罗马，后设置宋体，
已满足期刊的要求

并且支持mathtext.fontset，也就是数学公式文本字体的参数设置，
STIX 字体是专为科学出版设计的字体，但是中文和公式一起的时候中文还是无法显示，
如下面代码显示。需要在需要显示的地方单独设置fontname
"""

config = {
    "font.family" : ['Times New Roman', 'SimSun'],
    "mathtext.fontset": 'stix',
}

#更新全局属性配置
plt.rcParams.update(config)

x = [1, 2, 3, 4, 6.6, 8.4]
y = [random.randint(5,10) for i in range(6)]

fig = plt.figure()
#plt.xticks(x,['第{}个'.format(i) for i in range(6)], fontname = 'SimSun')
plt.xticks(x,['第{}个'.format(i) for i in range(6)])
#公式显示测试
#plt.xlabel('中文加公式测试 $\dfrac{t}{s}$')#这样写会出错
plt.xlabel('中文加公式测试 $\dfrac{t}{s}$', fontname = 'SimSun')
plt.plot(x,y)
plt.show()
```
