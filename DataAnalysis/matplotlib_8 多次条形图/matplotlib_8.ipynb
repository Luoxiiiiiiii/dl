```python
a=["猩球崛起3：\n终极之战", "敦刻尔克", "蜘蛛侠：\n英雄归来", "战狼2"]
b_16 =[15746,312,4497,319]
b_15 =[12357,156,2045,168]
b_14 =[2358,399,2358,362]
```

```python
from matplotlib import pyplot as plt
import matplotlib as mat

#可以都为双引号"
mat.rcParams["font.family"] = ["Times New Roman", "SimSun"]

#可以设置参数
bar_height = 0.1

#通过+0.1错开不同条形统计图
x_14 = list(range(len(a)))
x_15 = [i + bar_height for i in x_14]
x_16 = [i + bar_height for i in x_15]

fig = plt.figure(figsize = (10,5), dpi = 80)
plt.barh(x_14, b_14, height = bar_height, label = "9月14号", color = '#F2889B')
plt.barh(x_15, b_15, height = bar_height, label = "9月15号", color = '#95BFA4')
plt.barh(x_16, b_16, height = bar_height, label = "9月16号", color = '#F29E38')
#设置x和y的ticks的时候，x和y的选取是绝对的，不是跟随barh绘图函数中的参数
plt.xticks(fontsize = 15)
plt.yticks(x_15, a, fontsize = 15)
plt.legend(fontsize = 15)
plt.grid(alpha = 0.5)
plt.xlabel("票房", fontsize = 20)
plt.ylabel("")
plt.title("四部电影在三日的票房", fontsize = 20)
#savefig()必须放在show()之前!!
plt.savefig("./四部电影在三日的票房.png")
plt.show()
```
