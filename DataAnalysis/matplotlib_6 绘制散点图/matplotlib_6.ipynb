```python
"""
假设通过爬虫你获取到了北京2016年3,10月份每天白天的最高气温（分别位于列表a.b），那么此时如何寻找出气温和随时间（天）变化的某种规律？
39
拾初散点図
a=[11,17,16,11,12,11,12,6,6,7,8,9,12,15,14,17,18,21,16,17,20,14,15,15,15,19,21,22,22,22,23］
b=[26,26,28,19,21,17,16,19,18,20,20,19,22,23,17,20,21,20,22,15,11,15,5,13,17,10,11,13,12,13,6］
"""
```

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager

config = {
    "font.family" : ['Times New Roman', 'SimSun'],
    "mathtext.fontset" : 'stix',
}

plt.rcParams.update(config)

y_3 = [11,17,16,11,12,11,12,6,6,7,8,9,12,15,14,17,18,21,16,17,20,14,15,15,15,19,21,22,22,22,23]
y_10 = [26,26,28,19,21,17,16,19,18,20,20,19,22,23,17,20,21,20,22,15,11,15,5,13,17,10,11,13,12,13,6]
y_3_min = min(y_3)
y_3_max = max(y_3)
y_10_min = min(y_10)
y_10_max = max(y_10)
min_temp = min([y_3_min, y_10_min])
max_temp = max([y_3_max,y_10_max])



x_3 = range(1, 32)
x_10 = range(41, 72)

fig = plt.figure(figsize = (25,8), dpi = 80)

#scatter用来绘制散点图
plt.scatter(x_3, y_3, label = "3月份每天白天的最高气温")
plt.scatter(x_10, y_10, label = "10月份每天白天的最高气温")

_xticks_labels = ["3月{}号".format(i) for i in x_3]#这样表示更好
_xticks_labels += ["10月{}号".format(i-50) for i in x_10] 
#_xticks_labels = ["3月{}号".format(i) for i in range(1,32)]
#_xticks_labels += ["10月{}号".format(i) for i in range(1,32)] 

#是list(), 不是list[]
plt.xticks((list(x_3) + list(x_10))[::3], _xticks_labels[::3], fontsize = 15)
plt.yticks(range(min_temp, max_temp+1), fontsize = 15)

plt.xlabel("日期", fontsize = 20)
plt.ylabel("温度/摄氏度", fontsize = 20)
plt.title("3,10月份每天白天的最高气温", fontsize = 20)

plt.grid()
plt.legend(fontsize = 20)

plt.show()
```
