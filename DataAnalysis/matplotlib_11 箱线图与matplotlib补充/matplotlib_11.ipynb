```python
import matplotlib.pyplot as plt
import numpy as np

# 生成随机数据
# 使用固定的随机种子，确保结果可复现
np.random.seed(10)
# 生成100行3列的正态分布随机数
# 三个数据集的均值分别为3,5,4，标准差分别为1.25,1.00,1.25
data = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))
print(data)
# 创建图形
fig = plt.figure(figsize=(10,8), dpi=80)

# 绘制箱线图
# positions: 箱线图的位置
# widths: 箱体宽度
# patch_artist: 是否填充箱体
# showmeans: 是否显示均值（已设置为True）
# showfliers: 是否显示异常值（已设置为True）
# medianprops: 中位数线的样式设置（已改为红色）
# boxprops: 箱体的样式设置
# whiskerprops: 须线的样式设置
# capprops: 箱线图顶部和底部横线的样式设置

#默认将二维数组的每一列视为一个独立的数据集
plt.boxplot(
    data,
    positions=[2, 4, 6],
    widths=1.5,
    patch_artist=True,
    showmeans=True,
    showfliers=True,
    medianprops={"color": "red", "linewidth": 3.5},
    boxprops={"facecolor": "C1", "edgecolor": "white", "linewidth": 1.5},
    whiskerprops={"color": "C0", "linewidth": 11.5},
    capprops={"color": "C0", "linewidth": 1.5}
)

# 显示图形
plt.show()
```
