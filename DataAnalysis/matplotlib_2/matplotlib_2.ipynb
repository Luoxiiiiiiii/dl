```python
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(20,8),dpi=40)
#通过实例化一个figure并且传入参数，能够在后台使用该实例
#figsize是图片大小，dpi用来调整清晰度
#figsize=(20,8)是一个元组
x = range(2,26,2)
y = [14,13,15.5,23,5,23,21,21,14,25,14,30]
plt.plot(x,y)
plt.savefig("./fig_size.png")
#保存图片的方法，sig_size是图片名字，保存为png格式
#./表示是当前路径下的
#也可以保存为svg格式，矢量图
plt.show()
```

```python
x = range(2,26,2)
y = [14,13,15.5,23,5,23,21,21,14,25,14,30]

for i in x:
    print(i)
#range(x,y,z)，其中x包含y不包含，z是步长默认为1

plt.xticks(x)
#设置x的刻度,就是将x的每一个值显示在x轴上
#这个函数需要传入列表
plt.plot(x,y)
plt.show()
```

```python
x = range(2,26,2)
y = [14,13,15.5,23,5,23,21,21,14,25,14,30]

fig = plt.figure(figsize=(20,8))
#注意创建图形对象不能放在plt.xticks()后面
#因为plt.xticks()、plt.plot()、plt.xlabel()等都会在没有当前图像的情况下创建一个默认的图像
#而fig = plt.figure()是不管什么情况都主动创建了一个图像

xticks_label=[i/2 for i in range(4,49)]

#plt.xticks(xticks_label)
plt.xticks(xticks_label[::3])#列表取步长操作
#yticks方法同理
plt.yticks(range(min(y),max(y)+1))

#让x轴的刻度稀疏一些
#plt.xticks(range(2,26,0.5)) 这个是错误的，range里步长不能是float 

plt.plot(x,y)
plt.show()
```
