```python
from matplotlib import pyplot as plt

config = {
    "font.family" : ['Times New Roman', 'SimSun'],
    "mathtext.fontset" : 'stix',
}

plt.rcParams.update(config)
#添加\n能换行
a=["战狼2", "速度与激情8", "功夫瑜伽", "西游伏妖篇", "变形金刚5:\n最后的骑士", "摔跤吧!爸爸", "加勒比海盗5:\n死无对证", "金刚:骷髅岛", "极限特工:\n终极回归", "生化危机6:\n终章", "乘风破浪", "神偷奶爸3", "智取威虎山", "大闹天竺", "金刚狼3:\n殊死一战", "蜘蛛侠:\n英雄归来", "悟空传", "银河护卫队2", "情圣", "新木乃伊"]
b=[56.01, 26.94, 17.53, 16.49, 15.45, 12.96, 11.8, 11.61, 11.28, 11.12, 10.49, 10.3, 8.75, 7.55, 7.32, 6.99, 6.88, 6.86, 6.58, 6.23]

fig = plt.figure(figsize=(25, 8), dpi = 80)

plt.xticks(rotation = -90, fontsize = 15)
plt.yticks(fontsize = 15)
#bar用来画条形图
plt.bar(a, b, color = 'cyan', width = 0.3)

plt.ylabel("票房/亿元", fontsize = 20)
plt.show()
```

```python
#绘制横着的条形图

a=["战狼2", "速度与激情8", "功夫瑜伽", "西游伏妖篇", "变形金刚5:\n最后的骑士", "摔跤吧!爸爸", "加勒比海盗5:\n死无对证", "金刚:骷髅岛", "极限特工:\n终极回归", "生化危机6:\n终章", "乘风破浪", "神偷奶爸3", "智取威虎山", "大闹天竺", "金刚狼3:\n殊死一战", "蜘蛛侠:\n英雄归来", "悟空传", "银河护卫队2", "情圣", "新木乃伊"]
b=[56.01, 26.94, 17.53, 16.49, 15.45, 12.96, 11.8, 11.61, 11.28, 11.12, 10.49, 10.3, 8.75, 7.55, 7.32, 6.99, 6.88, 6.86, 6.58, 6.23]

fig = plt.figure(figsize=(15, 12), dpi = 80)

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
#bar用来画条形图,bar换成barh
plt.barh(a, b, color = 'cyan', height = 0.3)

plt.ylabel("票房/亿元", fontsize = 20)
plt.grid(alpha = 0.3)
plt.show()

#错误：got multiple values for argument经常遇到，给参数传了两个值
#在barh中，第二个参数 (b)是width
```
